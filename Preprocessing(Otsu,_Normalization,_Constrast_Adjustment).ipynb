```
import cv2
from google.colab.patches import cv2_imshow
# Load image, convert to grayscale, and find edges
from google.colab import drive
from google.colab import files
%matplotlib inline
from PIL import Image
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook
from matplotlib import pyplot as plt
from numpy import asarray

import numpy as np
from PIL import Image
import cv2

#drive.mount('/content/drive')

#import cv2
import os
import glob
img_dir = "/content/drive/My Drive/Kaggledataset/test/" # Enter Directory of all images 
data_path = os.path.join(img_dir,'*g')
print(data_path)
files = glob.glob(data_path)
data = []
for f1 in files:
  # img = cv2.imread(f1)
  # data.append(img)
#/Messidordataset/IMAGES/IMAGES/20051020_43882_0100_PP
  print(f1)
  #image = cv2.LoadImage(f1, CV_LOAD_IMAGE_COLOR)
  image = cv2.imread(f1)
  #print(image)
  if image is None:
    data.append(f1)
    #print(f1)
  #print(image.shape)
  #plt.imshow(ROI)
  #plt.show()
  #cv2.imshow('image', image)
  #print(image.size)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU + cv2.THRESH_BINARY)[1]
  # Find contour and sort by contour area
  cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  cnts = cnts[0] if len(cnts) == 2 else cnts[1]
  cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
  # Find bounding box and extract ROI
  for c in cnts:
      x,y,w,h = cv2.boundingRect(c)
      ROI = image[y:y+h, x:x+w, :]
      break
  ROI_new = cv2.cvtColor(ROI, cv2.COLOR_BGR2RGB)
  #plt.show()
  #normalized
  imgs_normalized = np.empty(ROI_new.shape)
  imgs_std = np.mean(ROI_new)
  imgs_mean = np.std(ROI_new)
  imgs_normalized = (ROI_new-imgs_std)/ imgs_mean
  for i in range(ROI_new.shape[0]):
      imgs_normalized[i] = ((imgs_normalized[i] - np.min(imgs_normalized[i])) / (np.max(imgs_normalized[i])-np.min(imgs_normalized[i])))*255
  clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(4,4))
  imgs_equalized = np.empty(imgs_normalized.shape)
  for i in range(imgs_normalized.shape[0]):
      imgs_equalized[i] = clahe.apply(np.array(imgs_normalized[i], dtype = np.uint8))
  invGamma = 1.0
  table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
  # apply gamma correction using the lookup table
  new_imgs = np.empty(imgs_equalized.shape)
  for i in range(imgs_equalized.shape[0]):
      new_imgs[i] = cv2.LUT(np.array(imgs_equalized[i], dtype = np.uint8), table)
  new_imgs = new_imgs/255.  #reduce to 0-1 range
  #print(new_imgs.getfield)
  #plt.imshow(new_imgs)
  #plt.show()
  #pyplot.imshow(pic)
  plt.imsave("/content/drive/My Drive/Kaggledataset/preprocessed/test/" + f1.split("/")[6], new_imgs)
  #new_im = Image.fromarray(new_imgs.astype('uint8'))
  #print(f1.split("/")[7].split(".")[0])
  #new_im.save("/content/drive/My Drive/Messidordataset/preprocessed/" + f1.split("/")[7])
  #break
print(data)
```
